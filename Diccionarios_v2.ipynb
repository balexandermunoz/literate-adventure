```
MyDictionaryToUse='/content/drive/Shared drives/Testimonios/DiccionariosVersionDic1_2020.csv'
Afects=Routines.DictsBuild(MyDictionaryToUse,Debug=False)
print(Afects.keys())
```

```
ReporteManual='/content/drive/Shared drives/Testimonios/ReporteTesteoManual.xlsx'
dicc = Routines.ConteoManual(ReporteManual,MyDictionaryToUse,'15_TM_HAP (2)',filtradas=True)
print(dicc)
```

```
#Ejemplo
ResultadoTextoRStem = Routines.contador_stemming(Routines.getText(FolderBase+TestimoniosNombres[0]), Afects)
print('\n Resultados por palabra: \n', ResultadoTextoRStem)
Metrica1=max(ResultadoTextoRStem, key=lambda key: sum(ResultadoTextoRStem[key]))
Metrica2=Routines.sumador(ResultadoTextoRStem)
print('Afectación con mayor ocurrencia: ', Metrica1)
print('Total de ocurrencias por afectación: \n', Metrica2)
```

```
ResultadosTestM1={}
ResultadosTestM2={}
ReporteManualDicc={}
for i in TestimoniosTestN:
  ResultadoT = Routines.contador_stemming(Routines.getText(FolderTestN+i), Afects)
  ResultadosTestM1[i]=max(ResultadoT, key=lambda key: sum(ResultadoT[key]))
  ResultadosTestM2[i]=Routines.sumador(ResultadoT)
  ReporteManualDicc[i]=Routines.ConteoManual(ReporteManual,MyDictionaryToUse,i.split(".")[0],filtradas=True)
for i in TestimoniosTestT:
  ResultadoT = Routines.contador_stemming(Routines.getText(FolderTestT+i), Afects)
  ResultadosTestM1[i]=max(ResultadoT, key=lambda key: sum(ResultadoT[key]))
  ResultadosTestM2[i]=Routines.sumador(ResultadoT)
  ReporteManualDicc[i]=Routines.ConteoManual(ReporteManual,MyDictionaryToUse,i.split(".")[0],filtradas=True)
```

```
import pprint
pprint.pprint(ReporteManualDicc)
```

```
ResultadosTestM2_perc={}
for key in ResultadosTestM2:
  ResultadosTestM2_perc[key]={item: value/sum(ResultadosTestM2[key].values()) for (item, value) in ResultadosTestM2[key].items()}
```

```
pprint.pprint(ResultadosTestM2_perc)
```

```
ReporteManualDicc_perc={}
for key in ReporteManualDicc:
  ReporteManualDicc_perc[key]={item: value/sum(ReporteManualDicc[key].values()) for (item, value) in ReporteManualDicc[key].items()}
pprint.pprint(ReporteManualDicc_perc)
```

```
ReporteMetrica2={}
for key in ReporteManualDicc_perc:
  Total=0
  for kkey in ResultadosTestM2_perc[key]:
    Total+=abs(ReporteManualDicc_perc[key].get(kkey,0)-ResultadosTestM2_perc[key].get(kkey,0))
  ReporteMetrica2[key]=Total/len(ResultadosTestM2_perc[key])
pprint.pprint(ReporteMetrica2)
```

```
ReporteMetrica3={}
for key in ReporteManualDicc_perc:
  DiffDicc={}
  for kkey in ResultadosTestM2_perc[key]:
    DiffDicc[kkey]=abs(ReporteManualDicc_perc[key].get(kkey,0)-ResultadosTestM2_perc[key].get(kkey,0))
  ReporteMetrica3[key]=DiffDicc
pprint.pprint(ReporteMetrica3)
```
