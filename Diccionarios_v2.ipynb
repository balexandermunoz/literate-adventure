```
ReporteManualDiccM2_perc={}
for key in ReporteManualDiccM2:
  ReporteManualDiccM2_perc[key]={item: value/sum(ReporteManualDiccM2[key].values()) for (item, value) in ReporteManualDiccM2[key].items()}
pprint.pprint(ReporteManualDiccM2_perc)
```

```
ReporteMetrica2={}
for key in ReporteManualDiccM2_perc:
  Total=0
  for kkey in ResultadosTestM2_perc[key]:
    Total+=abs(ReporteManualDiccM2_perc[key].get(kkey,0)-ResultadosTestM2_perc[key].get(kkey,0))
  ReporteMetrica2[key]=Total/len(ResultadosTestM2_perc[key])
pprint.pprint(ReporteMetrica2)
```

```
ReporteMetrica3={}
for key in ReporteManualDicc_perc:
  DiffDicc={}
  for kkey in ResultadosTestM2_perc[key]:
    DiffDicc[kkey]=abs(ReporteManualDicc_perc[key].get(kkey,0)-ResultadosTestM2_perc[key].get(kkey,0))
  ReporteMetrica3[key]=DiffDicc
pprint.pprint(ReporteMetrica3)
```
